```python
# Task 1: Merge and Sort Two Lists without Built-in Sort
def merge_and_sort(list1, list2):
    merged_list = list1 + list2
    # Implementing Bubble Sort
    for i in range(len(merged_list)):
        for j in range(0, len(merged_list) - i - 1):
            if merged_list[j] > merged_list[j + 1]:
                merged_list[j], merged_list[j + 1] = merged_list[j + 1], merged_list[j]
    return merged_list

# Example
list1 = [3, 1, 4]
list2 = [2, 5, 0]
print(merge_and_sort(list1, list2))  # Output: [0, 1, 2, 3, 4, 5]


# Task 2: Repeat Characters in a String
def repeat_characters(s, n):
    return ''.join([char * n for char in s])

# Example
s = "abc"
n = 3
print(repeat_characters(s, n))  # Output: "aaabbbccc"
```
