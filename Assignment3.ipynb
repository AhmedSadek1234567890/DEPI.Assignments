```python
# prompt: ### Question: Count Vowels in a String
# **Description:** Write a function named count_vowels that takes a string as an argument and returns the count of vowels (a, e, i, o, u) in the string.
# The function should be case-insensitive, counting both uppercase and lowercase vowels.
# '''
# # Example Input:
# text = "Hello World"
# # Expected Output:
# 3
# '''

def count_vowels(text):

  vowel_count = 0
  vowels = "aeiouAEIOU"
  for char in text:
    if char in vowels:
      vowel_count += 1
  return vowel_count

# Example usage
text = "Hello World"
vowel_count = count_vowels(text)
print(vowel_count)  # Output: 3
```

```python
# prompt: ### Question: Find the Maximum Value in a List Using a Function
# Description: Write a function named find_max that takes a list of integers as an argument and returns the maximum value in the list.
# Do not use Pythonâ€™s built-in max() function. The function should handle cases where the list is empty by returning None.
# '''
# # Example Input:
# numbers = [3, 5, 2, 8, 1]
# # Expected Output:
# 8
# '''

def find_max(numbers):

  if not numbers:
    return None

  max_value = numbers[0]
  for number in numbers:
    if number > max_value:
      max_value = number

  return max_value

# Example usage
numbers = [3, 5, 2, 8, 1]
max_number = find_max(numbers)
print(max_number)  # Output: 8
```
